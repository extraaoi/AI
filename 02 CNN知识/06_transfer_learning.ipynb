## 本节的内容是迁移学习，我们借用别人的网络模型和已经训练好的权重参数（或者是部分参数）来完成自己的任务。

```python
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
base_model = ResNet50(
    include_top = False, weights = 'imagenet', input_tensor = None,pooling = 'avg', classes = 10,
    classifier_activation = 'softmax')
base_model.summary()
```

### 注意看：Non-trainable params: 53,120；Trainable params: 23,534,592（可以让用户自己决定自己是否需要重新训练的权重参数）

## 二。我们来查看下模型中哪些层是可以训练的，哪些是不可以训练的

```python
for x in base_model.trainable_weights:
    print(x.name)
print('\n')
print('不可训练\n')
## 查看不可训练的层
for x in base_model.non_trainable_weights:
    print(x.name)
print('\n')
```

### 调整模型的层所有可训练，部分可训练
### 1. 全部可训练、全部不可训练
### 2. 部分可训练（前面某些层或者是后面某些层）

```python
##全部可训练
for layer in base_model.layers:
    layer.trainable = True
base_model.summary()
```

#### 注意看上面的可训练和布可训练的模型参数个数。理解他们的大概作用

```python
## 全部不可训练
for layer in base_model.layers:
    layer.trainable = False
base_model.summary()
```

### 模型可以部分训练。开启5层进行训练

```python
for layer in base_model.layers:
    layer.trainable = True
for layer in base_model.layers[0:-5]:
    layer.trainable = False
base_model.summary()
```

```python
## 查看下可训练的是不是最后5层？
for x in base_model.trainable_weights:
    print(x.name)
print('\n')
```

### 模型可以部分训练。开启8层进行训练?

```python
for layer in base_model.layers:
    layer.trainable = True
for layer in base_model.layers[0:-8]:
    layer.trainable = False
base_model.summary()
```

```python
## 查看下可训练的是不是最后5层？
for x in base_model.trainable_weights:
    print(x.name)
print('\n')
```

## 三。构建可训练模型
### 方法一。先冻结base_model,然后合并
### 方法二。合并以后再冻结

```python
num_classes  = 10
resnet50_fine_tune = Sequential()
resnet50_fine_tune.add(base_model)
resnet50_fine_tune.add(Dense(num_classes, activation = 'softmax'))
resnet50_fine_tune.compile(loss = 'categorical_crossentropy', optimizer='sgd',metrics=['accuracy'])
resnet50_fine_tune.summary()
```

#### 后面8层可训练，外加最后一个dense层需要训练

```python
num_classes  = 10
resnet50_fine_tune = Sequential()
resnet50_fine_tune.add(base_model)
resnet50_fine_tune.add(Dense(num_classes, activation = 'softmax'))
resnet50_fine_tune.layers[0].trainable = False
resnet50_fine_tune.compile(loss = 'categorical_crossentropy', optimizer='sgd',metrics=['accuracy'])
resnet50_fine_tune.summary()
```

#### 仅有最后一层开启训练
